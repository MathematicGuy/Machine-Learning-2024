## Preprocess

```python
import re
from nltk import word_tokenize, sent_tokenize
import string
from tqdm import tqdm

def tokenize(doc):
    tokens = word_tokenize(doc.lower())
    table = str.maketrans('', '', string.punctuation.replace("_", "")) #Remove all punctuation
    tokens = [w.translate(table) for w in tokens]
    tokens = [word for word in tokens if word]
    return tokens

full_data = ". ".join(full_data)
full_data = full_data.replace("\n", ". ")
corpus = []
sents = re.split(r'(?<=[^A-Z].[.?]) +(?=[A-Z])', full_data)
# sents = sent_tokenize(full_data)
for sent in tqdm(sents):
    corpus.append(tokenize(sent))
```

```python
vocab = set()
for sent in corpus:
    for word in sent:
        vocab.add(word)
vocab_list = list(vocab)
```

Get n-gram list where n is number of token for each iteration  

```python
def n_gram(tokens,n):
    """
    Input:
    tokens:List[str]: a list of tokens
    n: int: length of a n-gram

    Return:
    ngram_list: List[Tuple[str]]: a list of n-gram
    """
    # write your code here #
    ngram_list = []
    for i in range(0, len(tokens)+1):
      if len(tokens[i:i+n]) == n:
        ngram_list.append(tokens[i:i+n])
    return ngram_list
```

```python
# n_gram(sentence, n=2) -> n-gram word in a sentences
# n_gram(corpus, n=2) -> n-gram sentences in a corpus
o = 0
for sentence in corpus:
  print(n_gram(sentence, n=2))
  o += 1
  if o > 2:
    break
```

## 3.1 N-gram counts

Now we will need to store counts for each of the bigrams and trigrams in our text. Write a function called **get_ngram_counts**(corpus, n)
that takes your training corpus and uses your n_gram functions to calculate the n-gram counts within the training text.

```python
def get_ngram_counts(corpus, n):
    """
    Input:
    corpus: List[List[str]]: List of tokens in a sentence
    n: int: length of a n-gram

    Return:
    result: List[Dict[str: int]]: a list of i-gram (i in [1, n])
    each i-gram is a dictionary:
    key: i-gram
    value: number of occurrences of i-gram in the corpus
    """
    result = []
    # Write your code here
    for i in range(1, n+1): # 1,2,3
      n_gram_dict = {}
      for sentence in tqdm(corpus):
        ngram_list = n_gram(sentence, i)
        for ngram in ngram_list:
          ngram_tuple = tuple(ngram)
          if ngram_tuple in n_gram_dict:
            n_gram_dict[ngram_tuple] += 1
          else:
            n_gram_dict[ngram_tuple] = 1

      result.append(n_gram_dict)

    return result
```

### Knesser-Ney Smoothing (Only for bigram)

Stupid backoff and interpolation smoothing methods have one same problem: relying on MLE which, in some cases, does not reflect the true prob of a word. Knesser-Ney smoothing solve it by calulate the prob of a word based on the unique appearing of n-gram containing the word by the following fomula:

$
P_{\text{CONTINUATION}}(w) = \frac{\left| \{ w_{i-1} : c(w_{i-1}, w) > 0 \} \right|}
{\sum_{w'} \left| \{ w'_{i-1} : c(w'_{i-1}, w') > 0 \} \right|}
$

The prob of a word given the context (unigram), as a result, can be calculated by the following recursive formula:

$
P_{\text{KN}}(w_i \mid w_{i-1}) =
\frac{\max(c(w_{i-1}, w_i) - d, 0)}{c(w_{i-1})} + \lambda(w_{i-1}) P_{\text{CONTINUATION}}(w_i)
$

Where:
$
\lambda(w_{i-1}) = \frac{d}{c(w_{i-1})} \left| \{ w : c(w_{i-1}, w) > 0 \} \right|
$
