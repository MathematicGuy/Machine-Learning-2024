# Pie Chart
Use when you to see how each elements contribute to the whole.

```python
from matplotlib import pyplot as plt

plt.style.use("seaborn-v0_8-paper") # set style to xkcd
```

```python
slices = [30, 15, 25, 15, 15]
blue = '#00BFFF'
red = '#FF6347'
yellow = '#FFFF00'
green = '#00FF7F'

labels = ['Jungle', 'Mid', 'Baron', 'Dragon', 'Mix']
colors = [green, yellow, red, blue]
# represent how far a pie from the radius.
explode = [0, 0, 0, 0.2, 0] # explode mean exploded out

# autopct: auto percentiles
plt.pie(slices, labels=labels, explode=explode, startangle=130, autopct='%0.1f%%',wedgeprops={'edgecolor':'black'})
plt.title("Player by Lane in LoL")
plt.tight_layout()
plt.show()
```
