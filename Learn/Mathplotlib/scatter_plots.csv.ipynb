# Scatter Plots

Great when you want to show the relationship between 2 sets of value and see if they're correlated

Good for displaying trend and outliers

```python
import pandas as pd
from matplotlib import pyplot as plt
plt.style.use('seaborn-v0_8-poster')
```

```python
x = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6]
y = [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 8]

# color tensity level between 1 (lighter) and 10 (darker)
colors = [7, 5, 9, 7, 5, 7, 2, 5, 3, 7, 1, 2, 8, 1, 9, 2, 5, 6]
sizes = [209, 486, 381, 255, 191, 315, 185, 228, 174, 538, 239, 394, 399, 153, 273, 293, 436, 501]

#? s (size), c (color) can be a list of value corresponding to the data (each dots)
plt.scatter(x, y,
            s=sizes, # setting dots size
            # c = 'green', # setting dots color
            c=colors, # set dots color tensity level 
            cmap='Greens', # map green along with c color tensity level
            edgecolors='white', # dots edge color
            linewidths=1, # width of dots edge
            alpha=0.75 # transparency of dots
)

# add color bar
cbar = plt.colorbar() 
cbar.set_label('Satisfaction')

plt.show()
```

```python
data = pd.read_csv('data/yt_channel_data.csv') # data save from yt api
view_count = data['view_count']
likes = data['likes']
ratio = data['ratio']

plt.scatter(view_count, likes, ratio, c=ratio, cmap='summer', edgecolors='black', linewidths=1, alpha=0.75)

cbar=plt.colorbar()
cbar.set_label('Like / Dislike Ratio')

plt.title("Trending Youtube Video")

# normalize data display using log
plt.xscale('log')
plt.yscale('log')

plt.tight_layout()
plt.show()
```
