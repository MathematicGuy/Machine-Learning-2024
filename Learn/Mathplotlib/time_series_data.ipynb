# Time Series Data
Use to plot value over datetime. Like Stocks, Prices, Currency, etc..

```python
import pandas as pd
from datetime import datetime, timedelta
from matplotlib import pyplot as plt
from matplotlib import dates as mpl_dates
```

```python
plt.style.use('seaborn-v0_8')
```

```python
# 7 datetime value
dates = [
    datetime(2019, 5, 24),
    datetime(2019, 5, 23),
    datetime(2019, 5, 22),
    datetime(2019, 5, 21),
    datetime(2019, 5, 20),
    datetime(2019, 5, 19),
    datetime(2019, 5, 18),
    datetime(2019, 5, 17),
]
y = [0, 1, 3, 4, 6, 5, 7, 8]

plt.plot_date(dates, y, linestyle='solid')
plt.gcf().autofmt_xdate()

date_format = mpl_dates.DateFormatter('%b, %d %Y')

# change x-axis datetime format  
plt.gca().xaxis.set_major_formatter(date_format)
plt.tight_layout()
plt.show()
```

```python
data = pd.read_csv('data/bitcoin_price.csv')

data['Date'] = pd.to_datetime(data['Date']) # change date['Date'] from String to datetime datatype
data.sort_values('Date', inplace=True) # sort value from early to latest

dates = data['Date']
prices = data['Open']

# notice the data are just read as String and haven't recognize as data yet.
plt.plot_date(dates, prices, linestyle='solid')
plt.gcf().autofmt_xdate() # auto format datetime with value

date_format = mpl_dates.DateFormatter('%b, %d %Y') # set datetime format

# change x-axis datetime format  
plt.gca().xaxis.set_major_formatter(date_format)
plt.tight_layout()
plt.show()
```
