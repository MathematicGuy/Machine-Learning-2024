### Re-Scaling Data

```python
def Mean(X):
    m = X.shape[0]
    return np.sum(X)/m
# print(Mean(x_trainss))
```

```python
import math
def standard_deviation(X):
    mean = Mean(X)
    return np.sqrt(1/len(X) * sum(X - mean)**2)
```

```python
def normalization(X):
    X = (X - np.min(X)) / (np.max(X) - np.min(X))    
    return X  
```

```python
def zscore_standardization(X):
    mu = np.mean(X)
    sigma = standard_deviation(X)
    X = (X - mu)/sigma 
    return X
```

```python
X_normalized = normalization(x_train)
print(f'Normalization:\n{X_normalized[:5]}')

print()

X_standardized = zscore_standardization(x_train)
print(f'Z-Score Standardize:\n{X_standardized[:5]}')
```
