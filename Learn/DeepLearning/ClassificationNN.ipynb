### Activation Function

```python
def sigmoid(x):
    return 1 / 1 + np.exp(x)


def softmax(x):
    return np.exp(x) / sum(np.exp(x))
```

### Loss Function

```python
def CrossLossEntropy(y_true, y_pred):
    y_pred = np.clip(y_pred, 1e-15, 1 - 1e-15)
    CE = np.sum(-y_true*np.log(y_pred) - (1-y_true)*np.log(1-y_pred))
    return CE / len(y_true) 
```
