```python
import numpy as np
import matplotlib.pyplot as plt

# Sample TV Budget and Sales Data (for simplicity, small dataset)
TV_budget = np.array([50, 60, 70, 80, 90])
sales = np.array([20, 30, 40, 50, 60])

# Initialize parameters
w = 0  # weight
b = 0  # bias
learning_rate = 0.0001
epochs = 1000

# Number of data points
n = len(TV_budget)

# Gradient Descent
for epoch in range(epochs):
    # Predict the sales
    y_pred = w * TV_budget + b
    
    # Compute gradients
    dw = (-2/n) * np.sum(TV_budget * (sales - y_pred))
    db = (-2/n) * np.sum(sales - y_pred)
    
    # Update the parameters
    w -= learning_rate * dw
    b -= learning_rate * db
    
    # Optionally, print loss at intervals
    if epoch % 100 == 0:
        loss = np.mean((sales - y_pred) ** 2)
        print(f'Epoch {epoch}, Loss: {loss}, w: {w}, b: {b}')

# Final parameters after training
print(f"Final parameters: w = {w}, b = {b}")

# Predict sales for a new TV budget
new_budget = 100
predicted_sales = w * new_budget + b
print(f"Predicted sales for TV Budget = {new_budget}: {predicted_sales}")

# Plotting the result
plt.scatter(TV_budget, sales, color='blue', label='Actual Sales')
plt.plot(TV_budget, w * TV_budget + b, color='red', label='Regression Line')
plt.xlabel('TV Budget')
plt.ylabel('Sales')
plt.title('TV Budget vs Sales')
plt.legend()
plt.show()
```
